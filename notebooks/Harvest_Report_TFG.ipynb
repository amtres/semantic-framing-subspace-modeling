# TFG Harvest Report — Data Collection Summary
## Semantic Framing of Mental Health During COVID-19

**Period**: March 2020 – March 2021  
**Source**: GDELT API (Global Database of Events, Language, and Tone)  
**Strategy**: Day-by-media granular harvesting  
**Outlets**: 9 Spanish news publishers  

This notebook summarizes the data harvesting process and corpus statistics for the TFG research project.


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

# Monthly harvest progress data (from runs/corpus_progress_2020.md)
data = {
    'month': ['2020-03', '2020-04', '2020-05', '2020-06', '2020-07', '2020-08',
              '2020-09', '2020-10', '2020-11', '2020-12', '2021-01', '2021-02', '2021-03'],
    'records_saved': [2791, 8411, 5935, 4102, 3802, 3735, 3841, 2263, 2922, 3506, 4519, 3414, 4132],
    'mh_articles': [98, 463, 261, 169, 131, 141, 148, 73, 115, 126, 149, 119, 164],
    'mh_pct': [3.51, 5.50, 4.40, 4.12, 3.45, 3.78, 3.85, 3.23, 3.94, 3.59, 3.30, 3.49, 3.97],
    'la_vanguardia': ['ON', 'ON', 'OFF', 'OFF', 'OFF', 'ON', 'ON', 'ON', 'ON', 'ON', 'ON', 'ON', 'ON']
}

df = pd.DataFrame(data)
total_covid = df['records_saved'].sum()
total_mh = df['mh_articles'].sum()
avg_yield = df['mh_pct'].mean()
print("Total COVID articles harvested: {:,}".format(total_covid))
print("Total MH-filtered articles: {:,}".format(total_mh))
print("Average MH yield: {:.2f}%".format(avg_yield))
print()
df
```

## 1. Monthly Harvesting Volume

The bar chart below shows the total COVID-19 articles harvested per month (blue) and the mental health filtered subset (orange). The red markers indicate months where La Vanguardia was paused.


```python
fig, ax1 = plt.subplots(figsize=(14, 6))

x = np.arange(len(df))
width = 0.4

bars1 = ax1.bar(x - width/2, df['records_saved'], width, label='COVID-19 Articles', color='#2c3e50', alpha=0.7)
bars2 = ax1.bar(x + width/2, df['mh_articles'], width, label='MH-Filtered Articles', color='#e67e22', alpha=0.8)

# Mark La Vanguardia OFF months
for i, lv in enumerate(df['la_vanguardia']):
    if lv == 'OFF':
        ax1.scatter(i - width/2, df['records_saved'].iloc[i] + 200, marker='v', color='red', s=100, zorder=5)

ax1.set_xlabel('Month', fontsize=12)
ax1.set_ylabel('Number of Articles', fontsize=12)
ax1.set_title('Monthly Data Harvesting Volume', fontsize=14, fontweight='bold')
ax1.set_xticks(x)
ax1.set_xticklabels(df['month'], rotation=45, ha='right')
ax1.legend(loc='upper right')
ax1.grid(axis='y', alpha=0.3)

# MH% on secondary axis
ax2 = ax1.twinx()
ax2.plot(x, df['mh_pct'], color='#e74c3c', marker='o', linewidth=2, label='MH Yield %')
ax2.set_ylabel('MH Filter Yield (%)', fontsize=12, color='#e74c3c')
ax2.legend(loc='upper left')

plt.tight_layout()
plt.show()

print('Red markers = months where La Vanguardia was paused (harvesting instability)')
```

## 2. Pandemic Timeline Context

| Period | Event | Impact on Corpus |
|---|---|---|
| Mar 2020 | State of Alarm declared | Peak MH yield (5.5% in April) |
| May-Jul 2020 | La Vanguardia paused | Reduced coverage from one major outlet |
| Jun 2020 | "New Normal" begins | Gradual decline in volume |
| Sep 2020 | Second Wave onset | Return to school/work, volume rises |
| Oct 2020 | Selective confinements | Lowest harvest volume (2,263) |
| Jan 2021 | Third Wave + vaccine rollout | Volume recovery |


## 3. Corpus Summary

| Metric | Value |

| **Total COVID-19 articles** | 53,373 |
| **MH-filtered articles** | 2,157 |
| **Average MH yield** | 3.88% |
| **Peak MH month** | April 2020 (463 articles, 5.50%) |
| **Lowest MH month** | October 2020 (73 articles, 3.23%) |
| **Period** | March 2020 - March 2021 |
| **Publishers** | 9 Spanish outlets |
| **La Vanguardia gap** | May-July 2020 |

### Key Observations

1. **April 2020** had both the highest absolute volume and highest MH yield, reflecting the acute emotional shock of the lockdown.
2. **October 2020** had the lowest volume overall, coinciding with selective confinements and editorial fatigue.
3. The MH yield remained remarkably stable (3.2%-5.5%) across all months, suggesting mental health was a **persistent** editorial concern throughout the pandemic.
4. The La Vanguardia gap (May-July) is documented but does not affect corpus trends since the other 8 outlets maintained stable output.
